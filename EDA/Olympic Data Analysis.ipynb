```python
# create new dataframe merging sports and countries files
new_df=participants.merge(countries,how='left',on ='NOC')
new_df.head(10)
                
```

### handling missing data

```python
new_df.Height = new_df.Height.fillna(new_df.Height.mean())
new_df.Weight = new_df.Weight.fillna(new_df.Weight.mean())
new_df.Age= new_df.Age.fillna(new_df.Age.mean())
new_df.head(10)
```

```python
new_df.isnull().sum()
```

```python
TotalParticipants = participants[(participants.Sex=='F') | (participants.Sex=='M')][['Sex','Year']]
TotalParticipants =TotalParticipants.groupby('Year').count().reset_index()
TotalParticipants.head(10)
```

```python
plt.figure(figsize=(20, 10))
sns.barplot('Year', 'Sex', data=TotalParticipants,palette='pastel')
plt.title('Number of total participants over the years')
```

```python
#print winter games and year they were played
WinterGames = participants[(participants.Season=='Winter') ]
WinterGames =WinterGames.groupby('Year').count().reset_index()
plt.figure(figsize=(20, 10))
sns.barplot('Year', 'Season', data=WinterGames,palette='Set2')
plt.title('Winter Games over the years')
```
