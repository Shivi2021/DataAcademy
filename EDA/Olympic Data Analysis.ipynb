```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Import the dataset
participants = pd.read_csv("C:/TechTalent Academy/EDA project/sports.csv")
countries = pd.read_csv("C:/TechTalent Academy/EDA project/countries.csv")
```

```python
# create new dataframe merging sports and countries files
new_df=participants.merge(countries,how='left',on ='NOC')
new_df.head(10)
                
```

DATA CLEANING 

```python
#to check for any missing or null values
new_df.isnull().sum()
```

```python
#print team spain
new_df.query('Team_x =="Spain"').head(5)
```

```python
TotalParticipants = participants[(participants.Sex=='F') | (participants.Sex=='M')][['Sex','Year']]
TotalParticipants =TotalParticipants.groupby('Year').count().reset_index()
TotalParticipants
```

```python

#age distribution of the athletes
plt.figure(figsize =(12,6))
plt.title("Age distribution of athletes")
plt.xlable ="Age"
plt.ylabel ="Number of participants"
plt.hist(new_df.Age, bins =np.arange(10,80,2),color ="blue" ,edgecolor ="white");
```

```python
#List all unique sports
uniquesports= new_df.Sport.unique()
uniquesports
```

```python
Men = new_df[(new_df.Sex == 'M') & (new_df.Season == 'Summer')]
Women = new_df[(new_df.Sex == 'F') & (new_df.Season == 'Summer')]
```

```python
menweightlift =  Men.loc[Men['Sport'] == 'Weightlifting']
womenweightlift = Women.loc[Women['Sport'] == 'Weightlifting']
```

```python
plt.figure(figsize=(20, 10))
sns.pointplot('Year', 'Weight', data=menweightlift,palette='pastel')
plt.title('Weight of Men over the years')
```

```python
# Analyse participants data genderwise
gender_counts=new_df.Sex.value_counts()
gender_counts
```

```python
#Visualise the participation by gender
plt.figure(figsize=(12,6))
plt.title("Gender Distribution")
plt.pie(gender_counts,labels=gender_counts.index,autopct ='%1.1f%%',startangle=150,shadow= True)
```
