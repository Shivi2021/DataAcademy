```python
#Correlation
x=data['MaxHR']
y=data['HeartDisease']
correlate = x.corr(y)
correlate
```

```python
#Check for correlation amomg the data columns
correlation_data = data.corr()
correlation_data
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(correlation_data,annot= True,cmap='coolwarm')
```

From Above heaptmap for correlation between the different factors leading to heart disease.

Cross Tables

```python
newvar=pd.crosstab(data['Sex'],data['HeartDisease'])
print(newvar)
```

```python
newvar.plot(kind='bar',stacked=True,color=['blue','orange'],grid=False)
```

```python
newvar1=pd.crosstab(data['RestingECG'],data['HeartDisease'])
newvar1
```

```python
newvar1.plot(kind='barh',stacked=True,color=['red','yellow','green'])
```

```python
temp=pd.crosstab(index=data['HeartDisease'],columns=data['Sex'],margins=True)
temp
```

```python
temp.plot(kind='pie',subplots=True)
```
